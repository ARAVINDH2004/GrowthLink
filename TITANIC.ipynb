```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
data=pd.read_csv('tested.csv') #this is put the file to the variable
data.head() #to print the first 5 lines of the data
```

```python
data.isnull().sum() #this checks if there are any missing values in the entire dataset or dataframe.
#from the output we should see the 3 columns has missing data
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean()) #this will make sure , the missing value will be replaced with the mean of the entire column
data['Age'].isnull().sum() #this is to check if there is any missing values present in particular column
```

```python
data['Fare']=data['Fare'].fillna(data['Fare'].mean())
data['Fare'].isnull().sum()
```

```python
print(data['Cabin']) #from the output we can see that data type is object and out of 418 ,327 values are are NaN ,
#also our model needs to do the survival prediction.
#so i am gonna remove this entire column from the dataset
```

```python
data.drop(columns=['Cabin'],inplace=True) #removing the entire column
data.head() #checking if it is removed
```

```python
data.isnull().sum() #we could see there in no missing values in the dataset , so we can proceed for encode categorical part
```

```python
print(data['Sex'].unique())
print(data['Embarked'].unique())
#here we could the output, as the column "Sex" and "Embarked" contains categorical part,so now we can encode it
```

```python
encoder=LabelEncoder() #this will encode the value to zero's and one's as there are only two values in it, i have gone for it
data['Sex']=encoder.fit_transform(data['Sex']) #this will convert 'male' to 1 and 'Female' to 0.
print(data['Sex']) #Label Encoding
```

```python
mapping={'Q':0,'S':1,'C':2} #since there are three values i have done, ordinal encoding
data['Embarked']=data['Embarked'].map(mapping)
print(data['Embarked'])
```

```python
scaler=MinMaxScaler() #Since the "fare" value has high range, compressing their between 0 to 1 , will be much helpful for our predication model
data['Fare']=scaler.fit_transform(data[['Fare']]) #so, i have used MinMaxSclaer to compresss the value.
print(data['Fare'])
```

```python
data.head()
```

```python
data['Age']=scaler.fit_transform(data[['Age']]) #so, i have used MinMaxSclaer to compresss the value.
print(data['Age'])
```

```python
feature=data[['Sex','Age','SibSp','Parch','Fare']]#creating feature and target for classication purpose.
target=data['Survived']
```

```python
feature_train,feature_test,target_train,target_test=train_test_split(feature,target,test_size=0.2,random_state=42)#splititng the dataset into training and testing
```

```python
RFC_model=RandomForestClassifier(n_estimators=100, random_state=42)#using Random Forest Classifier to train the model
RFC_model.fit(feature_train,target_train)
```

```python
target_predict=RFC_model.predict(feature_test)#assigning the target predication
```

```python
accuracy = accuracy_score(target_test, target_predict)
precision = precision_score(target_test, target_predict, average='weighted')  #Evaluvating the model performance.
recall = recall_score(target_test, target_predict, average='weighted')
f1 = f1_score(target_test, target_predict, average='weighted')

# Print the results
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
```
